```
import os

fileNames = os.listdir('data_wo_sudir')
xls, txt, vcf, o, t = [], [], [], [], []
for fileName in fileNames:
  t.append(fileName)
  if fileName.endswith('.xls'):
    xls.append([len(fileName) - 4, fileName])
  elif fileName.endswith('.txt'):
    txt.append([len(fileName) - 4, fileName])
  elif fileName.endswith('.vcf'):
    vcf.append([len(fileName) - 4, fileName])
  else:
    o.append([len(fileName) - 4, fileName])

print('Total: {0}'.format(len(t)))
print('XLS: {0}'.format(len(xls)))
print('TXT: {0}'.format(len(txt)))
print('VCF: {0}'.format(len(vcf)))
print('Others: {0}'.format(len(o)))
```
